```python
#NLP preprocessing 
#Lower casing
#Removing Stop words
#Removing punctuations like . , ! $( ) * % @
#Removing URLs
#Tokenization
#Stemming
#Lemmatization
```

```python
import string
string.punctuation
```

```python
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree
df['clean_msg']= df['v2'].apply(lambda x:remove_punctuation(x))
df.head()
```
